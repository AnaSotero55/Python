```python
import numpy as np
import matplotlib.pyplot as plt

# Distribución normal
np.random.seed(19680801)

# media y desviación estándar
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=1, facecolor='y') # densidad, facecolor=color

plt.xlabel('Smarts') #nombre eje x
plt.ylabel('Probability') #nombre eje y
plt.title('Histogram of IQ') #textos
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') #dar posicion
plt.axis([40, 160, 0, 0.03])
plt.grid(True)# grid es la cuadricua
plt.show()
```
