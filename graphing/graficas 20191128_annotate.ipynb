# Annotate 20191128

```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('promedio', xy=(1, 1), xytext=(4, 1.7),
            arrowprops=dict(facecolor='red', shrink=0.05))

plt.ylim(-2,2)

plt.savefig("graph.png") #tambien se puede mandar guardar

plt.show()
```
