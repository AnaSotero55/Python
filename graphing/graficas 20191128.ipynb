# barh (bar horizontal)

```python
import numpy as np
import matplotlib.pyplot as plt

country = ("Alemania", "Espa√±a", "Francia", "Portugal")
position_y = np.arange(len(country))
population = (20, 45.56, 66.9, 10.32)

plt.barh(position_y, population, align = "center")

plt.yticks(position_y, country) #define la posicion de las barras 
plt.xlabel('Poblaci√≥n (millones)')
plt.title("Poblaci√≥n por pa√≠s")

plt.savefig("graph.png")
plt.show()
```

# bar vertical

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[72.81, 46.81, 52.4, 5], [79.43, 58.51, 57.25, 6], [82.67, 66.9, 65.64, 7]] #cada elemento del arreglo pertence al eje de las "x"
axisX = np.arange(4)
print(axisX+.3)
plt.bar(axisX + 0.0, data[0], color = "b", width = 0.25) 
plt.bar(axisX + 0.3, data[1], color = "c", width = 0.25)
plt.bar(axisX + 0.6, data[2], color = "m", width = 0.25)
plt.xticks(axisX+0.25, ["Alemania","Francia","Reino Unido", "Holanda"])


plt.savefig("graph.png")
plt.show()
```

# Pie

```python
# -*- coding: utf-8 -*-

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

values = ["Aprobados", "No aprobados", "deserci√≥n", "NP"]
data = [5, 20, 10, 30]
expl =(0.80, 0.05, 0, 0) #se da separacion a las partes de la grafica

fig, plot = plt.subplots()
plot.pie(data, explode=expl, labels=values, autopct='%1.1f%%', shadow=True, startangle=90) 

fig.savefig("graph.png")
plt.show()
```
