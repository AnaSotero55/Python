```python
import matplotlib.pyplot as plt
# %matplotlib_inline #para jupyter notebook
#Grafica 2 vectores
plt.plot(["Derecho_Inf","Informatica", "Mate", "Program", "Etica"],[9,8,9,8,10],'g-o') #Vector en Y 
                                                                                      #Vector en Y 
                                             #donde 'g-o' g=color en inglés y o=circulo, *=estrella, nada=linea por default
#Grafica por defecto el vector X = 1, 2, 3, 4
plt.axis([-1,5,7,11]) #-1 y 5 se refiere al eje de las 'x' y 7 y 11 al eje de las 'y' para centrar la grafica
plt.title('Promedio')
plt.ylabel('Calificacion')
plt.xlabel('Cantidad')
plt.show()
plt.savefig("graph.png")
```
